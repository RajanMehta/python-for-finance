```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# GSPC: S&P 500
# IXIC: NASDAQ
# erman DAX: GDAXI
# FTSE: London FTSE (Footsie)

data = pd.read_csv("Indices_Data_01.csv", index_col = "Date", parse_dates = ["Date"])
data.head(3)
```

```python
data.tail(3)
```

```python
((data/data.iloc[1])*100).plot(figsize=(15,6))  #because iloc[0] has NaN for GDAXI.
plt.show()
```

```python
ind_returns = (data / data.shift(1)) - 1
ind_returns.head(3)
```

```python
annual_ind_returns = ind_returns.mean() * 250
annual_ind_returns
```
